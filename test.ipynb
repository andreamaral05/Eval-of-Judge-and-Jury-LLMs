```python
from openai import OpenAI
 
# Inside HPC use the node name; with SSH tunnel use localhost
BASE_URL = "http://hpc02:11434/v1"   # or "http://localhost:11434/v1"
 
client = OpenAI(base_url=BASE_URL, api_key="ollama")
```

```python
question = "Explique rapidamente o que é overfitting."
resp = client.chat.completions.create(
    model="gemma3:4b",
    messages=[{"role": "user", "content": question}],
    temperature=0.2,
)

print(resp.choices[0].message.content)
```
